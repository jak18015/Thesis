```python
import os
import csv
```

```python
keep_list = []
with open('filelist.csv', 'r') as open_file:
    reader = csv.reader(open_file)
    for row in reader:
        keep_list.extend(row)

for item in keep_list:
    no_ext = item.split('.')[0]
    keep_list[keep_list.index(item)] = no_ext
keep_list
```

```python
absolute_file_paths = []
with open('frm2_imaging_paths.csv', 'r') as open_file:
    reader = csv.reader(open_file)
    for row in reader:
        absolute_file_paths.extend(row)
absolute_file_paths.remove('path')
absolute_file_paths[0]
```

```python
path_to_img_list = []
for path in absolute_file_paths:
    img_name = path.split('/')[-2] + '_' + path.split('/')[-1]
    img_name = img_name.split('.')[0]
    path_to_img_list.append(img_name)
path_to_img_list
```
