```python
import os
import pandas as pd
import re

wd = "C:/Users/jakek/OneDrive - University of Connecticut/1-Projects/frm2/"
os.chdir(wd)

csv_path = wd + "results/" + ".csv/" + "frm2-aid_apicoplast-inheritance-quantification.csv"

print(f'Current working directory: {os.getcwd()}')
print(f'CSV path: {csv_path}')
```

```python
csv = pd.read_csv(csv_path)
csv.head()
```

```python
ctrl = re.compile(".*ctrl.*")
kd = re.compile(".*kd.*")

csv.loc[csv['img'].str.contains('ctrl'), 'treatment'] = 'Control'
csv.loc[csv['img'].str.contains('kd'), 'treatment'] = 'FRM2-KD'

csv.head()
```

```python
csv.to_csv(csv_path, index=False)
```

```python
ctrl_counts = {}
kd_counts = {}

ctrl_counts['total'] = len(csv.loc[csv['treatment'] == 'Control'])
kd_counts['total'] = len(csv.loc[csv['treatment'] == 'FRM2-KD'])

ctrl_counts['defects'] = len(csv.loc[(csv['treatment'] == 'Control') & (csv['inheritanceDefect'] == 1)])
kd_counts['defects'] = len(csv.loc[(csv['treatment'] == 'FRM2-KD') & (csv['inheritanceDefect'] == 1)])

ctrl_defect_percent = ctrl_counts['defects'] / ctrl_counts['total'] * 100
kd_defect_percent = kd_counts['defects'] / kd_counts['total'] * 100

defect_notes = csv.loc[csv['inheritanceDefect'] == 1, ['img', 'note']]
defect_notes
```

```python
quote_from_tosetti_2019 = (
    "FRM2-KO parasites exhibited a defect in apicoplast inheritance (Figure 2F) ",
    "while the segregation of the Golgi was unaffected (Figure 2—figure supplement 2B). ",
    "The phenotype was partial; with about ~30% of vacuoles displaying a proper distribution ",
    "of apicoplast in all the parasites,~50% of the vacuoles showing at least one parasite ",
    "lacking the apicoplast, and ~20% with all the parasites lacking the apicoplast ",
    "(resulting from the reinvasion of parasites lacking the apicoplast) (Figure 2G)."
)

# Combine all the print statements into one
output = f"""
Percentages of images with an apicoplast inheritance defect.

Only images with a verifiable missing apicoplast were marked as having a defect.

Control: {ctrl_defect_percent:.0f}% ({ctrl_counts["defects"]}) defects among {ctrl_counts["total"]} images
FRM2-KD: {kd_defect_percent:.0f}% ({kd_counts["defects"]}) defects among {kd_counts["total"]} images

Defect notes:
"""

# Add defect notes
for i, row in defect_notes.iterrows():
    output += f'{row["img"]}: {row["note"]}\n'

# Add the quote
output += "\nquote from Tosetti et al. 2019:\n"
output += '"'
output += "\n".join(quote_from_tosetti_2019)
output += '"'

print(output)
```
